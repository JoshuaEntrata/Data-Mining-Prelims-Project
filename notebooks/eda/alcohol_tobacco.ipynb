#### **Importing Libraries**

```python
### Importing Libraries

# Data manipulation and analysis
import pandas as pd
import numpy as np

# Visualization
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Setup for the notebook
pd.set_option('display.max_columns', None)
%matplotlib inline
```

#### Load the preprocessed dataset

```python
# Loading the 2023 FIES Dataset Volume 2
final_fies = pd.read_csv('../../data/processed/final_fies.csv')
```

```python
final_fies
```

```python
# Calculate combined percentage of Tobacco + Alcohol
final_fies['percent_combined'] = (
    (final_fies['TOBACCO'] + final_fies['ALCOHOL']) / final_fies['NFOOD']
) * 100

# Group by social class and calculate mean percentages
percentages = final_fies.groupby('SOCIAL_CLASS')['percent_combined'].mean()

# Bar plot for better visualization of small percentages
plt.figure(figsize=(8, 5))
sns.barplot(
    x=percentages.index, 
    y=percentages.values, 
)

# Add percentage labels for clarity
for i, v in enumerate(percentages.values):
    plt.text(i, v + 0.1, f'{v:.2f}%', ha='center')

plt.title('Proportion of Tobacco & Alcohol in Non-Food Expenses by Social Class')
plt.xlabel('Social Class')
plt.ylabel('Percentage (%)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.ylim(0, max(percentages.values) + 1)  # Adjust y-axis for visibility
plt.show()
```
